Create a new database called 'crudeoilprice'

Connect to the 'master' database to run this snippet

```sql
USE master
```

Create the new database if it does not exist already

```sql
IF NOT EXISTS (
    SELECT [name]
        FROM sys.databases
        WHERE [name] = N'crudeoilprice'
)
CREATE DATABASE crudeoilprice
```

```sql
SELECT *
FROM crudeoilprice.dbo.CrudeOilProdPrice

-- Delete 2021 rows from table '[CrudeOilProdPrice]' in schema '[dbo]'
DELETE FROM crudeoilprice.dbo.CrudeOilProdPrice
WHERE [Year] = 2021
```

```sql


--- total crude oil production and total crude oil export in 2020 
SELECT SUM(Production) AS [total production 2020],
    SUM (Crude_Oil_Export) AS [total export 2020]
FROM crudeoilprice.dbo.CrudeOilProdPrice
WHERE [Year] = 2020

--- percentage of exported oil from produced in 2020
SELECT Month, Production AS [production 2020],
    Crude_Oil_Export AS [crude export 2020],
    (Crude_Oil_Export / Production) * 100  AS [% export per month 2020]
FROM crudeoilprice.dbo.CrudeOilProdPrice
WHERE Year = 2020
```

```sql
--- Year with the highest crude oil produced
SELECT DISTINCT (Year), SUM(Production)
  OVER (PARTITION BY Year) AS [cumulative total production],
  SUM(Crude_Oil_Export) OVER  (PARTITION BY Year) AS [total exported cruide oil]
 
FROM crudeoilprice.dbo.CrudeOilProdPrice
```

The table shown above is a representation of the cumulative total production of crude oil from from 2006 - 2021. The year with the highest oil production is 2010 with 29.6million barrels. according to the table shown above 2016 produced the lowest crude oil produced in Nigeria between 2006 and 2020.

```sql
---highest oil price.
SELECT Year, Month, max(Crude_Oil_Price) AS [Maximum Oil Price]
FROM crudeoilprice.dbo.CrudeOilProdPrice
GROUP BY Year, [Month]
ORDER BY max(Crude_Oil_Price) DESC
```

In June 2008 the price of oil was  $138.7 which appears to be the highest price of oil betwen 2006 and 2020. while April 2020 recorded the lowest prices of oil at $14.27.
